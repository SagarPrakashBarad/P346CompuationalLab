### Bisection method

Solve $f(x) = cos(x) âˆ’ x^3 = 0$ using bisection method, accurate up to 4 places in decimal.

```python
# defining the function
def f_1(x): 
    return math.cos(x) - x**3

# parameters
a = 0
b = 1
f_label = 'f(x) = \cos(x) - x^3'

plot_f_function(f_1, a, b, f_label)
```

```python
root = bisection(f_1, a, b, 1e-6)
print('Approximated root:', truncate_number(root, 4))
```

### False position Method

Solve $f(x) = 3x + sin(x) âˆ’ e^x = 0$ using regular falsi method, accurate up to 4 places in decimal.

```python
# defining the function
def f_1(x): 
    return 3*x + math.sin(x) - math.exp(x)

# parameters
a = 0
b = 2.1
f_label = 'f(x) = 3x + \sin(x) - e^x'

plot_f_function(f_1, a, b, f_label)
```

```python
root = regula_falsi(f_1, a, b, 1e-6)
print('Approximated root:', truncate_number(root, 4))
```

```python
# defining the function
def f_1(x): 
    return x*math.exp(x) - 2

def f_1_prime(x):
    return math.exp(x) + x*math.exp(x)

# parameters
a = -1.5
b = 1
f_label = 'f(x) = xe^x - 2'
f1_prime_label = "f'(x) = e^x + xe^x"

plot_functions(f_1, f_1_prime, a, b, f_label, f1_prime_label)
```

```python
x_0 = 0.1
root = newton_raphson(f_1, f_1_prime, x_0,  1e-6)
print('Approximated root:', truncate_number(root, 4))
```

```python
def f(x):
    return [
        x[0]**2 + x[1]**2 - 1,
        x[0]**2 - x[1] -1
    ]

def jacobian(x):
    return [
        [2*x[0], 2*x[1]],
        [2*x[0], -1]
    ]

x0 = [-0.5, 0.5]  # Initial guess
root_approximation = multivariate_newton_raphson(f, jacobian, x0, plot=True)
print("Approximated roots: ({}, {})".format(truncate_number(root_approximation[0], 4), truncate_number(root_approximation[1], 4)))
```
